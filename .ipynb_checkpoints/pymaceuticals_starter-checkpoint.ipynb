## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress
import numpy as np

# Study data files
mouse_metadata_path = "Mouse_metadata.csv"
study_results_path = "Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_master_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")
mouse_master_df.to_csv("newfile.csv",header=True)
# Display the data table for preview
mouse_master_df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumour volume for each regimen
regimen_group = mouse_master_df.groupby(["Drug Regimen"])

regimen_mean = regimen_group["Tumour Volume (mm3)"].mean()
regimen_median = regimen_group["Tumour Volume (mm3)"].median()
regimen_variance = regimen_group["Tumour Volume (mm3)"].var()
regimen_std = regimen_group["Tumour Volume (mm3)"].std()
regimen_sem = regimen_group["Tumour Volume (mm3)"].sem()

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumour volume. 
# Assemble the resulting series into a single summary dataframe.

summary_stats = pd.DataFrame({"Mean": regimen_mean, "Median":regimen_median, "Variance":regimen_variance, "Standard Deviation": regimen_std, "SEM": regimen_sem})

summary_stats
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumour volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
regimen_index = mouse_master_df.groupby('Drug Regimen')
summary_table = regimen_index.agg(['mean','median','var','std','sem'])["Tumour Volume (mm3)"]
summary_table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Group the dataframe by Mouse ID
grouped_df = mouse_master_df.groupby('Mouse ID')

# Use the .agg() method to find the maximum Timepoint for each mouse
max_timepoints = grouped_df['Timepoint'].agg('max')

# Reset the index of the resulting dataframe
max_timepoints = max_timepoints.reset_index()

# Merge the max_timepoints dataframe with the original dataframe
merged_df = pd.merge(mouse_master_df, max_timepoints, on=['Mouse ID', 'Timepoint'])
merged_df
```

```python
# Create a list that holds the treatment names as well as a second, empty list to hold the tumour volume data.
treatment_names = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
tumour_volumes = []

# Loop through each drug in the treatment list
for drug in treatment_names:
    # Locate the rows which contain mice on each drug and get the tumour volumes
    treatment_rows = merged_df[merged_df['Drug Regimen'] == drug]
    # Append the final tumour volumes for each drug to the tumour_volumes list
    tumour_volumes.append(treatment_rows['Tumour Volume (mm3)'].values[-1])

# Sort the tumour_volumes list
tumour_volumes.sort()

# Calculate the quartiles of the data set
q1 = tumour_volumes[int(len(tumour_volumes) * 0.25)]
q2 = tumour_volumes[int(len(tumour_volumes) * 0.5)]
q3 = tumour_volumes[int(len(tumour_volumes) * 0.75)]

# Calculate the lower bound
lower_bound = q1 - (1.5 * (q3 - q1))

# Calculate the upper bound
upper_bound = q3 + (1.5 * (q3 - q1))

# Determine which values are outliers
outliers = [x for x in tumour_volumes if x < lower_bound or x > upper_bound]

# Print the results
print("Treatment names: ", treatment_names)
print("Tumour volumes: ", tumour_volumes)
print("Lower bound: ", lower_bound)
print("Upper bound: ", upper_bound)
print("Outliers: ", outliers)
```

```python
# Generate a box plot of the final tumour volume of each mouse across four regimens of interest

# Create an empty list for each treatment group
capomulin_volumes = []
ramicane_volumes = []
infubinol_volumes = []
ceftamin_volumes = []

# Loop through each drug in the treatment list and populate the lists
for drug in treatment_names:
    # filter the dataframe by the treatment group
    treatment_group = merged_df.loc[merged_df['Drug Regimen'] == drug]
    #iterate over the rows of the filtered dataframe
    for i, row in treatment_group.iterrows():
        if drug == 'Capomulin':
            capomulin_volumes.append(row['Tumour Volume (mm3)'])
        elif drug == 'Ramicane':
            ramicane_volumes.append(row['Tumour Volume (mm3)'])
        elif drug == 'Infubinol':
            infubinol_volumes.append(row['Tumour Volume (mm3)'])
        elif drug == 'Ceftamin':
            ceftamin_volumes.append(row['Tumour Volume (mm3)'])

# Use the subplots function to create a figure with 4 subplots
fig, ax = plt.subplots(2, 2, figsize=(10, 6))
ax1, ax2, ax3, ax4 = ax.ravel()

# Create a box plot of the final tumour volumes for each treatment group on each subplot
ax1.boxplot(capomulin_volumes, flierprops={'marker': 'o', 'markerfacecolor': 'red', 'alpha': 0.5})
ax1.set_title('Capomulin')
ax2.boxplot(ramicane_volumes, flierprops={'marker': 'o', 'markerfacecolor': 'red', 'alpha': 0.5})
ax2.set_title('Ramicane')
ax3.boxplot(infubinol_volumes, flierprops={'marker': 'o', 'markerfacecolor': 'red', 'alpha': 0.5})
ax3.set_title('Infubinol')
ax4.boxplot(ceftamin_volumes, flierprops={'marker': 'o', 'markerfacecolor': 'red', 'alpha': 0.5})
ax4.set_title('Ceftamin')

# Add a title to the figure
fig.suptitle('Final Tumour Volume by Treatment Group')

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumour volume for the Capomulin regimen
capomulin_df = mouse_master_df.loc[mouse_master_df['Drug Regimen'] == 'Capomulin']
grouped_df = capomulin_df.groupby(['Mouse ID']).mean()
correlation = grouped_df['Weight (g)'].corr(grouped_df['Tumour Volume (mm3)'])

# get the x and y variable for the linear regression
x = grouped_df['Weight (g)']
y = grouped_df['Tumour Volume (mm3)']

# calculate the slope and y-intercept of the linear regression
slope, y_intercept = np.polyfit(x, y, 1)

# calculate the coefficient of determination (R-squared)
r_value, p_value = linregress(x, y)[:2]
r_squared = r_value**2

print(f'slope: {slope} y-intercept: {y_intercept} R-squared: {r_squared} correlation: {correlation}')
```

```python
# Plot the linear regression model on top of the previous scatter plot
# create the scatter plot
plt.scatter(x, y)

# plot the linear regression line
plt.plot(x, slope*x + y_intercept, color='red')

plt.xlabel('Weight (g)')
plt.ylabel('Tumour Volume (mm3)')
plt.title('Tumour volume vs Mouse weight for Capomulin')
plt.show()
```
